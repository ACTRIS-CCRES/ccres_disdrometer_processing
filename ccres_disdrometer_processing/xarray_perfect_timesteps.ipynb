```python
from pathlib import Path

import numpy as np
import pandas as pd
import xarray as xr
```

```python
d = pd.date_range('2023-01-01 14:00', '2023-01-01 18:00', freq="1min")
d
```

```python
ds = xr.Dataset(
    coords={
        "time": d
    },
    data_vars={
       "temp": (["time"], np.random.random(d.size))
    }
)

ds
```

```python
ds_res = ds.resample(time="1h").first()
```

```python
p_time = pd.date_range('2023-01-01 00:00', '2023-01-01 23:00', freq="1h")
p_time
```

```python
p_ds = xr.Dataset(
    coords={
        "time": p_time
    },
    data_vars={}
)

p_ds
```

```python
xr.merge([p_ds, ds_res])
```
